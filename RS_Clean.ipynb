```python
import json
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/clean_df.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df['year'] = df['year'].astype('Int64')

cs_matrix = np.load('https://drive.google.com/file/d/18UtsLYxbxtwqSrAZdLuC5JgmbaYN9dE4/view?usp=sharing')
cs = cs_matrix.f.arr_0

index = pd.Series(df['title'])
```

```python
def recommend(title:str) -> list:
  idx = index[index == title].index[0]
  score = pd.Series(cs[idx]).sort_values(ascending=False)
  top_50 = list(score.iloc[1:10].index)

  recommendations = df.iloc[top_50].to_json(orient='records')

  return recommendations
```

```python
res = recommend('Cowboy Bebop')

json.loads(res)
```
